**Objetivo:** receber um arquivo em XLSX e convertê-lo em CSV (para download) e em objeto Datasets (para upload no Hugging Face). O modelo baixado localmente pode ser utilizado na realização de testes. O modelo enviado ao Hugging Face pode ser utilizado no treinamento de modelos posteriormente.

**Informações técnicas:**
- O arquivo original contém as colunas: `Perguntas`, `Respostas` e `Contexto`;
- São adicionadas as colunas: `instruction`, `output` e `input` para substituir as colunas originais, respectivamente;
- As colunas originais são apagadas, e o arquivo é convertido;
- O download do CSV é feito localmente e o arquivo sofre upload para o `Hugging Face`.

```python
# Faça o login no HuggingFace usando o token do mesmo, para que o dataset convertido seja gravado no seu space

import os
from huggingface_hub import login
hf_token = os.getenv("HF_TOKEN")
login(token=hf_token) #é só clicar no botão Login sem precisar digitar o Token
```

# Formatação do Dataset

```python
# Passo 0: importação de bibliotecas

import pandas as pd
from datasets import Dataset
from google.colab import files

# Passo 1: Fazer upload do arquivo Excel e inserir nome do arquivo
uploaded = files.upload()
#nome_arquivo = '2024-03-19-QeA-MMGD-GeradasIA.xlsx'
nome_arquivo = '2024-05-31-QeA-MMGD-OriginaisSite.xlsx'
# Passo 2: Carregar o arquivo Excel usando pandas
df = pd.read_excel(nome_arquivo, sheet_name="QeA")

# Passo 3: criar colunas "output", "input" e "instruction" baseadas nas colunas que já estão no arquivo
df["output"] = df.Resposta
df["input"] = df.Contexto
df["instruction"] = df.Pergunta

# Passo 4: apagar colunas "Pergunta", "Resposta" e "Contexto"

df.drop(columns=["Pergunta", "Resposta", "Contexto"], inplace=True)

# Passo 5: transformar o arquivo resultante em CSV e fazer download local para verificação (opcional)
caminho_arquivo_csv = 'MMGD_Alpaca.csv'
df.to_csv(caminho_arquivo_csv, index=False, encoding='utf-8')

files.download(caminho_arquivo_csv)

# Converter o arquivo já convertido (df) em um Dataset do Hugging Face
hf_dataset = Dataset.from_pandas(df)

# Faça o upload do dataset para o Hugging Face
# Substitua "nome_do_seu_dataset" pelo nome que você quer dar ao seu dataset
hf_dataset.push_to_hub("2024-05-31-QeA-MMGD-OriginaisSite")
```
